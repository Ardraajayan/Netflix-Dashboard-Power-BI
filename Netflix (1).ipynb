Find Maximum & Minimum Duration Movies


*This can be determined by seperating the category into Movies and TV Shows.Then replace the string part from the data and change its data type.*

```python
mv =df.loc[(df['Category'] =='Movie')]
```

```python
mv.Duration= mv.Duration.apply(lambda x:x.replace("min","") if 'min' in x else x)
mv.head(2)
```

```python
mv.loc[:,["Duration"]]=mv.loc[:,["Duration"]].apply(lambda x: x.astype('int64',errors='ignore'))
mv.describe()
```

```python
shortest_movie = mv.loc[mv['Duration']==np.min(mv.Duration)]
shortest_movie
```

```python
longest_movie = mv.loc[mv['Duration']==np.max(mv.Duration)]
longest_movie
```

```python
show =df.loc[(df['Category'] =='TV Show')]
```

```python
show.Duration= show.Duration.apply(lambda x:x.replace("Season","") if 'Season' in x else x)
show.head(2)
```

```python
show.Duration= show.Duration.apply(lambda x:x.replace("s","") if 's' in x else x)
show.head(2)
```

```python
show.loc[:,["Duration"]]=show.loc[:,["Duration"]].apply(lambda x: x.astype('int64',errors='ignore'))
show.describe()
```

```python
longest_show = show.loc[show['Duration']==np.max(show.Duration)]
longest_show
```

```python
shortest_show = show.loc[show['Duration']==np.min(show.Duration)]
shortest_show.head(1)
```
